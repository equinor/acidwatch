```python
from acidwatch import Client
import pandas as pd

session = Client()
```

```python
session.list_models()
```

```python
concs = {
    "O2": 30,
    "H2O": 30,
    "H2S": 0,
    "SO2": 10,
    "NO2": 20,
}
params = {}
T = 200  # K
P = 10  # bar a
```

```python
session.run_model("tocomo", concs, params, temperature=T, pressure=T)
```

```python
dfs = []
for value in range(10, 50, 5):
    df = session.run_model(
        "phpitz", {**concs, "H2O": value}, params, temperature=T, pressure=T
    )
    df.set_index([pd.Series([value], name="H2O[in]")], inplace=True)
    dfs.append(df)
df = pd.concat(dfs)
df
```
